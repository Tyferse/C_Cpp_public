```python
f = open("pos.txt", 'r').read()
pos = [[]]
iter = f.split('\n\n', 1)[0]
for line in iter.split('\n'):
    pos[0].append([])
    for p in list(map(float, line.split(' '))):
        pos[0][-1].append(p)

for iter in f.split('\n\n')[1:]:
    pos.append([[] for _ in range(len(pos[0]))])
    for line in iter.split('\n')[1:]:
        if line.isspace() or not line:
            continue

        j = int(line.split(' ', 1)[0])
        for p in list(map(float, line.split(' ')[1:])):
            pos[-1][j].append(p)

print(pos)
```

```python
import matplotlib.pyplot as plt


def gen_frame(iteration, i, lims):
    fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
    fig.set_figwidth(8)
    fig.set_figheight(8)

    # ax.set_box_aspect([4,4,4])
    # ax.set_facecolor('white')
    # ax.autoscale(enable=True, axis='both')
    plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
    ax.view_init(elev=45, azim=45)
    ax.dist = 6.9

    ax.set_xlim(lims[0][0], lims[0][1])
    ax.set_ylim(lims[1][0], lims[1][1])
    ax.set_zlim(lims[2][0], lims[2][1])
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    ax.scatter([p[0] for p in iteration], [p[1] for p in iteration], [p[2] for p in iteration], s=[400 for _ in range(len(iteration))])

    plt.savefig(f'frames/{i:0>4}.png')
    ax.clear()
    plt.close()

lims = [[], [], []]
lims[0] = [min([p[0] for iteration in pos for p in iteration]), max([p[0] for iteration in pos for p in iteration])]
lims[1] = [min([p[1] for iteration in pos for p in iteration]), max([p[1] for iteration in pos for p in iteration])]
lims[2] = [min([p[2] for iteration in pos for p in iteration]), max([p[2] for iteration in pos for p in iteration])]
gen_frame(pos[0], 0, lims)
```

```python
for i, it in enumerate(pos):
    gen_frame(it, i, lims)
```

```python
import cv2
import os


frame_width = 800
frame_height = 800
fps = 30

# Путь для сохранения видео
output_video_path = 'nbodies.mp4'

# Инициализация объекта видео
out = cv2.VideoWriter(output_video_path,
                      cv2.VideoWriter.fourcc(*'mp4v'),  # mp4v
                      fps, (frame_width, frame_height))

# Папка с кадрами
frames_folder = "frames"

# Считывание кадров и запись в видео
for frame_name in os.listdir(frames_folder):
    frame_path = os.path.join(frames_folder, frame_name)
    frame = cv2.imread(frame_path)
    out.write(frame)

out.release()
cv2.destroyAllWindows()
```
